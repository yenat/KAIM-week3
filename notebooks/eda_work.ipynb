```python
# Cell 1: Import necessary libraries and the preprocessing script
import sys
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Add the parent directory to the Python path
sys.path.insert(0, os.path.abspath(os.path.join(os.getcwd(), '..')))
from src.txt_to_csv_clean import convert_txt_to_csv, clean_csv
```

```python
#  Load the cleaned data
df = pd.read_csv('/path/to/your/data')
df.head(10)
print(df.shape)

```

```python
# Data Summarization - Descriptive Statistics
desc_stats = df.describe()
desc_stats
```

```python
# Data Structure
data_types = df.dtypes
data_types
```

```python
# Univariate Analysis - Distribution of Numerical Variables
numerical_columns = ['TotalPremium', 'TotalClaims']
for col in numerical_columns:
    # Display key statistics
    print(f"Statistics for {col}:")
    print(df[col].describe())
    print("\n")
    
    # Plot histogram
    plt.figure(figsize=(10, 6))
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()


```

```python
# Univariate Analysis - Distribution of Categorical Variables
categorical_columns = ['CoverType', 'VehicleType', 'PostalCode']
for col in categorical_columns:
    # Display counts for each category
    print(f"Counts for {col}:")
    print(df[col].value_counts())
    print("\n")
    
    # Plot bar chart
    plt.figure(figsize=(10, 6))
    sns.countplot(x=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Bivariate/Multivariate Analysis
# Correlations and Associations

# Calculate correlation between TotalPremium and TotalClaims
correlation = df[['TotalPremium', 'TotalClaims']].corr()
print("Correlation between TotalPremium and TotalClaims:")
print(correlation)
print("\n")

# Exploring relationships between TotalPremium and TotalClaims
plt.figure(figsize=(12, 8))
sns.scatterplot(x='TotalPremium', y='TotalClaims', hue='PostalCode', data=df)
plt.title('Relationship between TotalPremium and TotalClaims by PostalCode')
plt.xlabel('TotalPremium')
plt.ylabel('TotalClaims')
plt.legend(title='PostalCode', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Correlation Matrix
correlation_matrix = df[['TotalPremium', 'TotalClaims']].corr()
print("Correlation Matrix:")
print(correlation_matrix)
print("\n")

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Data Comparison
# Trends Over Geography

# Display key statistics for TotalPremium by PostalCode
print("Summary statistics for TotalPremium by PostalCode:")
print(df.groupby('PostalCode')['TotalPremium'].describe())
print("\n")

plt.figure(figsize=(12, 8))
sns.boxplot(x='PostalCode', y='TotalPremium', data=df)
plt.title('Change in TotalPremium by PostalCode')
plt.xlabel('PostalCode')
plt.ylabel('TotalPremium')
plt.xticks(rotation=45)
plt.show()
```

```python
# Data Comparison
# Change in TotalPremium by VehicleType

# Display key statistics for TotalPremium by VehicleType
print("Summary statistics for TotalPremium by VehicleType:")
print(df.groupby('VehicleType')['TotalPremium'].describe())
print("\n")

plt.figure(figsize=(12, 8))
sns.boxplot(x='VehicleType', y='TotalPremium', data=df)
plt.title('Change in TotalPremium by VehicleType')
plt.xlabel('VehicleType')
plt.ylabel('TotalPremium')
plt.xticks(rotation=45)
plt.show()
```

```python
# Outlier Detection for TotalClaims

# Display key statistics for TotalClaims to identify potential outliers
print("Summary statistics for TotalClaims:")
print(df['TotalClaims'].describe())
print("\n")

plt.figure(figsize=(10, 6))
sns.boxplot(x='TotalClaims', data=df)
plt.title('Box Plot for TotalClaims')
plt.xlabel('TotalClaims')
plt.show()

```

```python
# Distribution of TotalPremium by CoverType

# Display key statistics for TotalPremium by CoverType
print("Summary statistics for TotalPremium by CoverType:")
print(df.groupby('CoverType')['TotalPremium'].describe())
print("\n")

plt.figure(figsize=(12, 8))
sns.violinplot(x='CoverType', y='TotalPremium', data=df)
plt.title('Distribution of TotalPremium by CoverType')
plt.xlabel('CoverType')
plt.ylabel('TotalPremium')
plt.xticks(rotation=45)
plt.show()
```

```python
# Relationship between TotalPremium and TotalClaims

# Display key statistics for TotalPremium and TotalClaims
print("Summary statistics for TotalPremium:")
print(df['TotalPremium'].describe())
print("\n")

print("Summary statistics for TotalClaims:")
print(df['TotalClaims'].describe())
print("\n")

# Calculate correlation between TotalPremium and TotalClaims
correlation = df[['TotalPremium', 'TotalClaims']].corr()
print("Correlation between TotalPremium and TotalClaims:")
print(correlation)
print("\n")

plt.figure(figsize=(12, 8))
sns.jointplot(x='TotalPremium', y='TotalClaims', kind='reg', data=df)
plt.title('Relationship between TotalPremium and TotalClaims')
plt.xlabel('TotalPremium')
plt.ylabel('TotalClaims')
plt.show()
```

```python
# Distribution of TotalClaims by VehicleType

# Display key statistics for TotalClaims by VehicleType
print("Summary statistics for TotalClaims by VehicleType:")
print(df.groupby('VehicleType')['TotalClaims'].describe())
print("\n")

plt.figure(figsize=(12, 8))
sns.boxplot(x='VehicleType', y='TotalClaims', data=df)
plt.title('Distribution of TotalClaims by VehicleType')
plt.xlabel('VehicleType')
plt.ylabel('TotalClaims')
plt.xticks(rotation=45)
plt.show()
```
